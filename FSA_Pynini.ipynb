```python
import pynini
import pywrapfst
```

```python
isym = pynini.SymbolTable()
isym.add_symbol("#", 0)
for c in range(97, 97+26):
  isym.add_symbol(chr(c), c)

cap = pynini.accep("cap")
cap.set_input_symbols(isym)
cap
```

```python
cat = pynini.accep("cat")
cat.set_input_symbols(isym)
cat
```

```python
captain = pynini.accep("captain")
captain.set_input_symbols(isym)
captain
```

```python
fsa = cap.union(cat.union(captain))
fsa.set_input_symbols(isym)
fsa
```

```python
fsa.optimize()
fsa
```
